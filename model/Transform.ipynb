# Transform

```python
from glob import glob
import pandas as pd
```

```python
glob('../data/*.csv')
```

```python
df_cases = pd.read_csv('../data\\cases.csv', index_col=0)
df_mysj = pd.read_csv('../data\\mysj.csv', index_col=0)
df_vaccine = pd.read_csv('../data\\vaccine.csv', index_col=0)
df_population = pd.read_csv('../data\\population.csv', index_col=0)
```

```python
df_cases
```

```python
df_cases = df_cases[['date', 'cases_new', 'state_id', 'cases_recovered', 'cases_death', 'cases_active']]
```

```python
df_mysj
```

```python
df_mysj = df_mysj[['date', 'state_id', 'checkins', 'unique_ind']]
```

```python
df_vaccine
```

```python
df_vaccine = df_vaccine[['date', 'state_id', 'cumul_full']]
```

```python
df_population = df_population[['state_id', 'pop']]
df_population
```

```python
df_vaccine = pd.merge(df_vaccine, df_population, on=['state_id'])
df_vaccine['percent_vax'] = round(df_vaccine['cumul_full']/df_vaccine['pop'],2)
```

```python
df_vaccine
```

```python
df_train = pd.merge(df_cases, df_mysj, on=['date', 'state_id'])
df_train = pd.merge(df_train, df_vaccine, on=['date', 'state_id'])
df_train
```

```python
df_train.to_csv('train.csv')
```
